### Multiplication
- Vector by Vector (dot product)
u · v = u₁v₁ + u₂v₂ + u₃v₃ + … + uₙvₙ
- Matrix by Vector
- Matrix by Matrix

```python
u.shape[0]
```

```python
v.shape[0]
```

```python
u
```

```python
v
```

```python
def vector_vector_multi(u, v):
    assert u.shape[0] == v.shape[0] # to perform vector multiplication must have the same size

    n = u.shape[0]
    result = 0.0

    for i in range(n):
        result = result + u[i] * v[i]

    return result
```

```python
vector_vector_multi(u, v)
```

```python
u.dot(v)
```
